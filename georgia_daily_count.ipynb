```python
lag = 14

date = 200

count = []
for day in range(date):
    if day < date-lag:
        day_count = data[day]
    else:
        day_count = data[day]*(date-day)/lag
        
    count.append(day_count)
    
count = np.array(count)

plt.figure(figsize=(13,7))
plt.plot(data)
plt.plot(count)
plt.vlines(date-lag,0,count.max())
plt.show()
```
